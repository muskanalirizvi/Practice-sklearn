```python
import pandas as pd
import numpy as np


np.random.seed(42)

data = {
    'Education_Level': np.random.choice(['High School', 'Bachelors', 'Masters', 'PhD'], 100),
    'Experience_Level': np.random.choice(['Entry', 'Mid', 'Senior'], 100),
    'Job_Title': np.random.choice(['Engineer', 'Doctor', 'Artist', 'Lawyer', 'Scientist'], 100)
}

df = pd.DataFrame(data)

df.head(20)


```

```python
df.groupby('Education_Level').count()
```

```python
df.groupby('Job_Title').count()
```

```python
df.groupby('Experience_Level').count()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['Job_Title'], axis = 1)
y = df['Job_Title']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder

```

```python
oe = OrdinalEncoder(categories = [['High School', 'Bachelors', 'Masters', 'PhD'],['Entry', 'Mid', 'Senior']])
```

```python
oe.fit(x_train)
x_train_encoded = oe.transform(x_train)
x_test_encoded = oe.transform(x_test)
x_train_encoded
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(y_train)
```

```python
y_train
```

```python
y_train_encoded = le.transform(y_train)
y_test_encoded = le.transform(y_test)
y_train_encoded
```
