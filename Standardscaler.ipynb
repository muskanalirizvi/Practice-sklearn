```python
import pandas as pd
```

```python
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset('tips')

# Duplicate the data to have at least 1000 rows
extended_tips = pd.concat([tips]*((1000 // len(tips)) + 1), ignore_index=True)

# Trim the dataframe to exactly 1000 rows
df = extended_tips.head(1000)

# Print the first few rows of the dataframe
df.head()

```

```python
df.shape
```

```python
df = df.drop(['sex','smoker','day','time'], axis = 1)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(df)
```

```python
df_scaled = scaler.transform(df)
df_scaled = pd.DataFrame(df_scaled, columns = df.columns)
```

```python
df_scaled
```
