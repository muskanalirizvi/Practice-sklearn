```python
import pandas as pd

df = pd.read_csv('cars.csv')
df.head()
```

```python
ValueCount = df['brand'].value_counts()
LessFrequencyBrands = ValueCount[ValueCount<100].index
LessFrequencyBrands
```

```python
df['brand'] = df['brand'].replace(LessFrequencyBrands,'other')
```

```python
df.tail(20)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(['selling_price'],axis=1),df['selling_price'] ,test_size=0.3)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
ohe = OneHotEncoder(drop = 'first', sparse_output = False)

x_train_encoded = ohe.fit(x_train[['brand', 'fuel', 'owner']])


```

```python
x_train_encoded = ohe.transform(x_train[['brand', 'fuel', 'owner']])
x_test_encoded = ohe.transform(x_test[['brand', 'fuel', 'owner']])
```

```python
column_names = ohe.get_feature_names_out(['brand', 'fuel','owner'])
```

```python
x_train_encode = pd.DataFrame(x_train_encoded, columns = column_names)
x_test_encode = pd.DataFrame(x_test_encoded, columns = column_names)
```

```python
x_train_encode.head()
```

```python
x_train_Dataframe = pd.concat([x_train.reset_index(drop=True).drop(['brand', 'fuel', 'owner'], axis=1), x_train_encode], axis=1)
columns_to_encode = ['brand', 'fuel', 'owner']
x_test_Dataframe = pd.concat([x_test.reset_index(drop=True).drop(['brand', 'fuel', 'owner'], axis=1), x_test_encode], axis=1)
```
