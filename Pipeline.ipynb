```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df = df[['survived', 'sex',	'age', 'fare' , 'embarked', 'deck']]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('survived',axis=1),df['survived'],test_size=0.3,random_state=42)
```

```python
x_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
trf1 = ColumnTransformer([
    ('siage', SimpleImputer(), [1]),
    ('sideck', SimpleImputer(strategy = 'most_frequent'), [4]),
    ('siembarked', SimpleImputer(strategy = 'most_frequent'), [3])
], remainder = 'passthrough')
```

```python
trf2 = ColumnTransformer([
    ('ohe_sex_deck', OneHotEncoder(sparse_output = False, drop = 'first'),[0,4]),
], remainder = 'passthrough')
```

```python
trf3 = ColumnTransformer([
    ('scaler', StandardScaler(),slice(0,14)),
], remainder = 'passthrough')
```

```python
pipe = Pipeline([
    ('trf1', trf1),
    ('trf2', trf2),
    ('trf3', trf3)
])
```

```python
pipe.fit(x_train,y_train)
```
